```python
## 1. Data Loading
```

```python
import pandas as pd
import matplotlib.pyplot as plt

smoking_df = pd.read_csv("../data/smoking_cancer_dataset.csv")
youth_df = pd.read_csv("../data/youth_smoking_drug_data_10000_rows_expanded.csv")
```

```python
## 2. Data Overview
```

```python
smoking_df.info()
smoking_df.head()

youth_df.info()
youth_df.head()
```

```python
## 3. Data Cleaning
Basic data cleaning was performed, including handling missing values, standardizing text fields, and checking for duplicate records.
```

```python
smoking_df = smoking_df.dropna()
youth_df = youth_df.dropna()

smoking_df = smoking_df.apply(
    lambda col: col.str.strip().str.title() if col.dtype == "object" else col
)

youth_df = youth_df.apply(
    lambda col: col.str.strip().str.title() if col.dtype == "object" else col
)
```

```python
## 4. Adult Smoking Analysis
This section explores smoking patterns among adults, focusing on smoking status, intensity, duration, and respiratory symptoms.
```

```python
smoking_df['Smoking_Status'].value_counts().plot(kind='bar')
plt.title('Smoking Status Distribution Among Adults')
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.show()
```

```python
(
    smoking_df
    .groupby('Smoking_Status')['Chronic_Cough']
    .value_counts(normalize=True)
    .unstack()
    .plot(kind='bar')
)

plt.title('Proportion of Chronic Cough by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Proportion')
plt.show()
```

```python
## 5. Youth Smoking Analysis
This section examines smoking prevalence among youth and explores potential influencing factors such as age group, peer influence, and access to preventive resources.
```

```python
youth_df.groupby('Age_Group')['Smoking_Prevalence'].mean().plot(kind='bar')
plt.title('Average Smoking Prevalence by Age Group (Youth)')
plt.xlabel('Age Group')
plt.ylabel('Smoking Prevalence (%)')
plt.show()
```

```python
## 6. Key Insights

- Former smokers show higher average smoking intensity and longer smoking duration compared to current smokers, suggesting prolonged exposure prior to cessation.
- Respiratory symptoms such as chronic cough are more prevalent among current and former smokers than among never smokers.
- Youth smoking prevalence varies across age groups, indicating that smoking behavior begins early and increases with age.
- Aggregated youth-level factors such as peer influence, school programs, and access to counseling show only marginal differences in smoking prevalence, highlighting the multifactorial nature of smoking initiation.
- While short-term correlations among youth appear weak, adult data reveals clearer long-term health consequences associated with prolonged smoking exposure.
```
