```python
import pandas as pd

smoking_df = pd.read_csv("smoking_cancer_dataset.csv")
youth_df = pd.read_csv("youth_smoking_drug_data_10000_rows_expanded.csv")

smoking_df.head(), youth_df.head()
```

```python
smoking_df = smoking_df.dropna()
youth_df = youth_df.dropna()
```

```python
smoking_df = smoking_df.apply(
    lambda col: col.str.strip().str.title() if col.dtype == "object" else col
)

youth_df = youth_df.apply(
    lambda col: col.str.strip().str.title() if col.dtype == "object" else col
)
```

```python
cough_by_smoking = (
    smoking_df
    .groupby('Smoking_Status')['Chronic_Cough']
    .value_counts(normalize=True)
    .rename('Proportion')
    .reset_index()
)

cough_by_smoking
```

```python
cough_yes = cough_by_smoking[cough_by_smoking['Chronic_Cough'] == 'Yes']
cough_yes
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.bar(
    cough_yes['Smoking_Status'],
    cough_yes['Proportion']
)
plt.title("Proportion of Chronic Cough by Smoking Status")
plt.xlabel("Smoking Status")
plt.ylabel("Proportion")
plt.show()
```
